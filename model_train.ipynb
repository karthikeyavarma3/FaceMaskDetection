```python
import tensorflow as tf
import numpy as np
import cv2
import os
from tensorflow import keras
import time
from PIL import Image
```

```python
size = 224
```

```python
parent_dir = r'C:\Projects\Dataset\original_set'

categories = ['Facemask' , 'incorrect' , 'no_face_mask']
```

```python
training_data = []
```

```python
for category in categories:
    path = os.path.join(parent_dir , category)
    label = categories.index(category)
    for img in os.listdir(path):
        image = cv2.imread(os.path.join(path , img))
        image_rgb = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
        img_size = cv2.resize(image_rgb , (size , size))
        training_data.append([img_size , label])
```

```python
import random

random.shuffle(training_data)
```

```python
x = []
y = []
```

```python
for features , label in training_data:
    x.append(features)
    y.append(label)
```

```python
X = np.array(x).reshape(-1 , size , size , 3)
Y = np.array(y)
```

```python
print(Y[0] , X[0])
```

```python
X = X/255.0
```

```python
print(X[0])
```

```python
import pickle

pickle_out = open('x.pickle' , 'wb')
pickle.dump(X , pickle_out)
pickle_out.close()

pickle_out = open('y.pickle' , 'wb')
pickle.dump(Y , pickle_out)
pickle_out.close()
```

```python
import pickle

pickle_in = open("x.pickle" , 'rb')
X = pickle.load(pickle_in)

pickle_in = open("y.pickle" , 'rb')
Y = pickle.load(pickle_in)
```

```python
print(Y[-1] , X[0])
```

```python
from keras import layers

from keras.losses import SparseCategoricalCrossentropy
from keras.applications import MobileNet
```

```python
model = MobileNet(weights='imagenet' , include_top=False , input_shape=(224,224,3))
```

```python
model.summary()
```

```python
base_input = model.layers[0].input
```

```python
base_output = model.layers[-1].output
```

```python
flat_layer = keras.layers.Flatten()(base_output)
final_output = keras.layers.Dense(128 , activation='relu' , name = 'my_layer_1')(flat_layer)
final_layer = keras.layers.Dense(3 , activation='softmax' , name = 'fully_connected_layer')(final_output)
```

```python
new_version = keras.Model(inputs = base_input , outputs = final_layer)
```

```python
new_version.save('untrained_v2')
```

```python
new_version.summary()
```

```python
new_version.compile(optimizer='adam' , loss = SparseCategoricalCrossentropy() , metrics = ['accuracy'])
```

```python
from keras.callbacks import Callback
from keras.callbacks import ModelCheckpoint
from keras.callbacks import EarlyStopping
```

```python
check_point = ModelCheckpoint(
    filepath = 'best_model.v2',
    monitor = 'val_accuracy',
    save_best_only=True,
    save_weights_only=False,
    mode='max',
    verbose=1
)
```

```python
early_stopping = EarlyStopping(
    monitor='val_accuracy',
    patience=2,
    mode='max',
    verbose=1
)
```

```python
for layer in new_version.layers[1:10]:
    layer.trainable = False

for layer in new_version.layers[10:]:
    layer.trainable = True
```

```python
new_version.summary()
```

```python
new_version.fit( X, Y , epochs = 5 , validation_split = 0.1  , batch_size = 64 , callbacks=[check_point , early_stopping])
```

```python
new_version.save('model_to_use_mobilenet_97')
```
